# Galaxy samples comparison for clustering measurements in photometric surveys 

## A case study using DESY3 Data

We explore the constraining power of different galaxy samples from the DESY3 data. Samples are selected through a series of cuts in the data, mainly magnitude (flux limits) and color cuts. Our indicator will be the forecasted mutual forecasted errors on constraining the parameters $\Omega_{m,0} - \sigma_8$.

### Theory

Let's start by briefly explaining the theory and the formulas we are going to use. 

###### ANGULAR POWER SPECTRA:

Using the Limber  and flat-sky approximations (valid for $\ell > 10$)  the angular power spectra of the projected field in two redshift bins $i$ and $j$ as: 

\begin{equation} 
C_\ell^{ij} = \int_0^\infty dz \frac{H(z)}{c}\frac{W^i(z)W^j(z)}{\chi(z)^2}P_{NL}\left(k=\frac{\ell+1/2}{\chi(z)}, z \right).
\end{equation}

Here, $H(z)$ is the Hubble parameter at redshift $z$ and $P_{NL}(k,z)$ is the non-linear matter power spectrum at wavenumber $k$ and redshift $z$. In the above expression we have not included the effects of redshift space distortions. These can be shown to be negligible in the angular scales under consideration. 

 The weighting kernel for angular clustering in a redshift bin $i$ is given by:
 
\begin{equation}
W^{i}(z) = b(z)\frac{dn^i}{dz},
\end{equation}

where $dn^i/dz$ is the normalized redshift distribution of galaxies in that bin. The bias factor, $b(z)$ that accounts for the fact that the observed galaxy overdensity field is a biased tracer of the underlying matter overdensity field. We assume a linear a scale independent bias factor, i.e. that the two fields are related as $\delta_g (z) = b(z) \delta_m(z)$.

###### FORECASTING FORMALISM:

To forecast constraints we rely on the standard Fisher recipe. The Fisher matrix provides an approximation for the covariance of the parameters a future experiment will try to measure. 

In the case where our data vector is the observed angular power spectra (including the cross correlations) it can be shown that it takes the form:

\begin{equation}
F_{\mu \nu} = \frac{f_{sky}}{2} \sum_{\ell} (2\ell + 1)\mbox{Tr}\left(\mathbf{\hat{C}}_\ell^{-1} \mathbf{\hat{C}}_{\ell,\mu} \mathbf{\hat{C}}_\ell^{-1} \mathbf{\hat{C}}_{\ell,\nu} \right)
\end{equation}

where $,_\mu \equiv \frac{\partial}{\partial \theta_\mu}$.  $\mathbf{\hat{C}}$ is a matrix with elements the observed power spectra $\hat{C}_\ell^{ij}$:

\begin{equation}
\hat{C}_\ell^{ij} = C_\ell^{ij} + \delta^{ij}\frac{1}{\bar{n}^i},
\end{equation}
where $\bar{n}^i$ is the number of galaxies per steradian in the $i$-th redshift bin.

In the above formula we take into account all the cross-correlations between redshift bins. In the case where we consider only the auto-correlations, the formula reduces to:

\begin{equation}
F_{\mu \nu} = \sum_i \sum_{\ell_{min}}^{\ell_{max,i}} \frac{1}{\sigma^2_{\ell,i}}\frac{\partial C_\ell^{i}}{\partial \theta_\mu} \frac{\partial C_\ell^{i}}{\partial \theta_\nu},
\end{equation}

where:

\begin{equation}
\sigma_{\ell,i} = 
\delta C_\ell = \sqrt{\frac{2}{f_{sky}(2\ell + 1)}}\left(C_\ell^{i} + \frac{1}{\bar{n}^i} \right).
\end{equation}

The outer sum is over the redshift bins. The maximum $\ell$ at each bin is given by $\ell_{max} = k_{ql}\chi(z_{med})$, where $k_{ql} = 0.2 \,h \,\mbox{Mpc}^{-1}$, the quasilinear scale.

 A commonly used metric to measure the ability of survey/sample under given specifications to constrain cosmology is the figure of merit (FoM). For a subset (of the total number) of cosmological parameters $\mathbf{\theta}$ we want to constrain, this can be calculated:
 
\begin{equation}
\mbox{FoM}_\mathbf{\theta} = \left[\det\left[ (\mathbf{F}^{-1})_\mathbf{\theta}\right]\right]^{-1/2}
\end{equation}

In the case where this subset of cosmological parameters coincides with the set of cosmological parameters we leave free to vary, the above reduces to:

\begin{equation}
\mbox{FoM} = \sqrt{\det{\mathbf{F}}}
\end{equation}

An intuitive way to understand the figure of merit, is two consider the marginalized posterior of the joint constraints of two parameters. Then, the figure of merit is inversly proportional to the area of the confidence ellipse of the constraints on these two parameters.


##### WHY DIFFERENT SAMPLES? SYSTEMATICS vs STATISTICS

The case behind using different samples for clustering measuremetns is the tradeoffs between the statistical power of a sample and the systematics. 

Statistical power of photometric surveys comes from covered area, to reduce cosmic variance, and from the number density of galax- ies, to reduce noise contributions when estimating summary statis- tics. Maximizing the number density of galaxies requires the inclu- sion of faint, small, and poorly understood galaxies, which give rise to additional systematics. The trade-off between statistical power and systematics needs to be simulated carefully to select optimal galaxy samples and to focus future research on the most limiting factors of an analysis (Krause & Eifler COSMOLIKE paper).

Here our focus is the trade-off between number (density) of galaxies used and photo-z errors.


##### SAMPLES CONSIDERED 


In DES SV and DES Y1 analyses the galaxy sample used for clustering meauserements is that selected through the RedMaGic algorithm, an algorithm that selects Luminous Red Galaxies (LRGs). The photometric redshifts of these galaxies are excellent ($\sigma_z/(1+z) < 0.017$ in DES Y1), but it is sparse ($6.6 \times 10^5$ galaxies). 

Here we explore if we can use larger samples in Y3 analysis to get better constraints, having in mind these trade-offs between statistical power and photo-zs. 

We consider and compare three ategories of samples:

$\bullet$ FULL FLUX LIMITED SAMPLE: 

After performing some quality cuts (for star-galaxy separation) and magnitude cuts that ensure that the sample is complete  up to the flux limit, we forecast its constraining power.

$\bullet$ COLOR LIMITED SAMPLE:

Then, we further perform cuts based on color, in order to select a sample that it dominated by red galaxies, but contains more galaxies than the LRGs selected through the RedMaGic algorithm. We follow the procedure described in the BAO sample selection paper (detailed description later). 

$\bullet$ REDMAGIC SAMPLE:

Finally, for completeness and direct comparison, we consider a RedMaGic sample, the sample that conventionally used for clustering measurements in DES.



##### PROCEDURE/COMPARISONS

Comparing different samples in a fair way is probably not an easy task. Different samples have i) Different redshift distributions. For example, RedMaGic galaxies have a sharp cut at $z=0.9$, while when considering flux limited/color limited samples we can go deeper. ii) Different samples have different photo-z errors, and thus the optimal binning is (?) different, as well as the importance of including cross-correlations can be very important when the bin width is comparable to the redshift error.

We will start with the simplest possible comparison and we will gradually explore more complicated cases/ further comparisons.

So, initially we will consider, as in the DES Y1 clustering analysis five redshift bins in the interval $z \in [0.15,0.9]$ for all samples.

In our first comparison we do not consider cross-correlation between the bins, but then we also explore how including them change the forecasted constraints.


The fiducial cosmology used is the best fit $\Lambda$CDM cosmology : DES Y1 + Planck + JLA + BAO:

\begin{equation}
\Omega_{m} = 0.301, \quad \sigma_8 = 0.798, \quad n_s = 0.973, \quad \Omega_b = 0.048, \quad h = 0.682 
\end{equation}

for the bias we consider a fiducial constant bias of $b = 1.6$ for redMaGic and color-cut samples and $b = 1.0$ for the full sample. 

Bias is certainly  topic we have to explore more.

At first we fix all parameters except $\Omega_m, \sigma_8$ to their fiducial values. Then we can leave more parameters free and compare how results change.

Calculate Fisher matrix and forecast constraints for such a sample.

```python
from scipy.interpolate import UnivariateSpline as spl
#============================================================
bias = 1.6
f_sky = 4580.0/41253.0

nz = 1000 #number of steps to use for the radial/redshift integration
zarray = np.linspace(0,4.0,nz)
z = zarray[1:-1]

Fisher_FL = 0.0 #initialize matrix 
for k in range(0,5):
    if (k==0):
        bin_z = bin_1
    elif (k==1):
        bin_z = bin_2
    elif (k==2):
        bin_z = bin_3
    elif (k==3):
        bin_z = bin_4
    else:
        bin_z = bin_5
        
    mean_z = np.mean(bin_z)
    N_gal = np.float(np.size(bin_z))
    
    y_like, x_like = np.histogram(bin_z, bins=200, normed=True) # take a histogram for the redshift distribution 
    x_like = x_like[:-1]
            
    bin_sp = spl(x_like,y_like,s=0.0, ext=1)     #Interpolate 
    n_z = bin_sp(z)   # this is the normalized n(z) in the range (0.0,4.0)
    
    Fisher_FL += FSB(mean_z, bias, n_z, f_sky, N_gal)
```

### Color cuts


Next we want to select a red sample for clustering measurements, optimized such that the $\Omega_m - \sigma_8$ plane FoM is maximum. 

Similarly to the BAO selection paper and the DES Y1 explorations, we impose a color cut of the form:

\begin{equation}
(i_{auto} - z_{auto}) + a_1(r_{auto} - i_{auto}) > a_2
\end{equation}

We want to run a grid of $[a_1,a_2]$ values, define a sample based on color cuts for each pair of values in the grid, calcuate redshift distributions and number of galaxies in each redshift bin and then calculate the Fisher matrix and thus the forecasted figure of merit.

For $a_1$ and $a_2$ let's choose:

\begin{equation} 
a_1 \in [0.5-2.0]
\end{equation}

\begin{equation}
a_2 \in [0.5 - 3.0]
\end{equation}

In another notebook we have shown that assuming a much larger grid does not affect the results.

```python
#============================================================
bias = 1.6
f_sky = 4580.0/41253.0

#a_1 range for the grid
a_1_gr = np.linspace(0.45,2.05,43)
#a_2 range for the grid
a_2_gr = np.linspace(0.44,3.05,43)


size = np.size(a_1_gr)  # size of a_1_gr has to be the same as the size of a_2_gr
#============================================================
# Initialize matrix
FoM = np.zeros([size,size])

for i in range(0,size):
    a_1_t = a_1_gr[i]
    for j in range(0,size):
        a_2_t = a_2_gr[j]
        
        
        #Now apply to the sample the cuts for these values of a_1, and a_2
        z_mean_temp = z_mean_1[(((i_auto_1-z_auto_1)+a_1_t*(r_auto_1-i_auto_1))>a_2_t)]
        z_mc_temp = z_mc_1[(((i_auto_1-z_auto_1)+a_1_t*(r_auto_1-i_auto_1))>a_2_t)]
        
        
        Fish_mat  = 0.0 # initalize fisher matrix
        #Create five bins and calculate Fisher matrix in each of them
        for k in range(0,5):
            z_i = 0.15 + k*0.15  # lower redshift limit of the bin
            z_f = 0.15 + (k+1)*0.15  # upper redshift limit of the bin
            
            bin_z = z_mc_temp[(z_mean_temp>z_i)&(z_mean_temp<z_f)]  #select the galaxies with mean redshifts inside the phot-z limits 
           
            if (np.size(bin_z)>0.0):
                mean_z = np.mean(bin_z)   #mean redshift of the bin
                N_gal = 64.0*np.size(bin_z)  #number of galaxies in the bin. note: here multiplied by 64
                #because i have used a sample which is 1/64 of the total
            
                y_like, x_like = np.histogram(bin_z, bins=200, normed=True) # take a histogram for the redshift distribution 
                x_like = x_like[:-1]
            
                bin_sp = spl(x_like,y_like,s=0.0, ext=1)     #Interpolate 
                n_z = bin_sp(z)   # this is the normalized n(z) in the range (0.0,4.0)
            
                Fish_mat += FSB(mean_z, bias, n_z, f_sky, N_gal)
            else:
                Fish_mat += np.array([[1.0e-20, 1.0e-20], [1.0e-20, 1.0e-20]])
            
        
        
        FoM[i][j] = np.sqrt(np.linalg.det(Fish_mat))
            
            
        
FoM_matr = np.matrix.transpose(FoM)
FoM_matr = FoM_matr/np.max(FoM_matr)
```

```python
import scipy.ndimage
def FoM_return(a_1,a_2):
    FoMino = FoM_matr
    return FoMino

plot_pretty() 
fig = plt.figure(figsize = (6,6))

a_1, a_2 = np.meshgrid(a_1_gr, a_2_gr)
FoM_pr = FoM_return(a_1, a_2)

FoM_pr = scipy.ndimage.gaussian_filter(FoM_pr, 1)
levels = np.linspace(0.0,1.0,200)
image = plt.contourf(a_1, a_2, FoM_pr , levels=levels, cmap=plt.cm.hot_r)
C = plt.contour(a_1, a_2, FoM_pr, 10,  colors=('k', 'k', 'k', 'k','k', 'whitesmoke', 'w', 'w','w','w'), linestyle ='--',  linewidth=.5)
plt.clabel(C, inline=2, fontsize=8)

plt.xlabel("$a_1$", fontsize=15);plt.ylabel("$a_2$", fontsize=15)


plt.xlim(0.5,2.0);plt.ylim(0.5,3.0)

cbar = plt.colorbar(image, orientation='vertical')
cbar.set_label('FoM (normalized)', fontsize=14)

plt.savefig('FoM_colorcuts.pdf')
#plt.savefig('FoM_colorcuts.png')
plt.show() 
```

### Explore color cuts based on the above plot

### RedMaGic Galaxies

Now let's compare the forecasted constraints from the RedMaGic sample for DES Y3, the sample conventionally used for clustering measurements.

Actually there are three separate redMaGiC samples  generated from the e1 data, referred to as the high-density, high-luminosity, and higher-luminosity samples. The corresponding luminosity thresholds and comoving densities for these samples are, respectively, $L_{min} = 0.5L_$ , $L_$ , and $1.5L_$ , and $\bar{n} = 10^{3}$, $4  10^{4}$, and $10^{4}$ galaxies/$(h^{1} Mpc)^3$, where $h  H0/(100 km sec^{1} Mpc^{1})$ parametrizes the Hubble constant.

These galaxies are placed in five nominally disjoint redshift bins. The low- est three bins $z = [(0.15  0.3), (0.3  0.45), (0.45  0.6)]$ are high-density, while the galaxies in the two highest redshift bins $(0.6  0.75)$ and $(0.75  0.9)$ are high-luminosity and higher-luminosity, respectively (DES Y1 results, arXiv: 1708.01530).

### Binning and redshift distributions

We will bin the RedMaGic galaxies in 5 redshift bins between $z = 0.15$ and $z = 0.9$. We will use the high-dens sample (red_1) for the first three bins, the high luminosity sample (red_2) for the fourth one and the higher luminosity sample (red_3) for the last bin (0.75-0.9). 

In order to produce the "true" redshift redshift distribution in each bin, we assign to each galaxy a random number (redshift) drawn from a gaussian with mean the best fit redshift $z_{RM}$ and standard deviation $\sigma_z$ (unique for each galaxy). 

```python
from numpy.random import normal as gauss
true_z_1 = gauss(red_photo_z_1,red_error_z_1)
true_z_2 = gauss(red_photo_z_2,red_error_z_2)
true_z_3 = gauss(red_photo_z_3,red_error_z_3)
```

```python
bin_1_rm = true_z_1[(red_photo_z_1>0.15)&(red_photo_z_1<0.30)] 
bin_2_rm = true_z_1[(red_photo_z_1>0.30)&(red_photo_z_1<0.45)]
bin_3_rm = true_z_1[(red_photo_z_1>0.45)&(red_photo_z_1<0.60)]
bin_4_rm = true_z_2[(red_photo_z_2>0.60)&(red_photo_z_2<0.75)]
bin_5_rm = true_z_3[(red_photo_z_3>0.75)&(red_photo_z_3<0.90)]
```

#### Cosmological constraints

Use the above results to forecast cosmological constraints from the redmagic sample

```python
Fisher_RM = 0.0 #initialize matrix 
for k in range(0,5):
    if (k==0):
        bin_z = bin_1_rm
    elif (k==1):
        bin_z = bin_2_rm
    elif (k==2):
        bin_z = bin_3_rm
    elif (k==3):
        bin_z = bin_4_rm
    else:
        bin_z = bin_5_rm
        
    mean_z = np.mean(bin_z)
    N_gal = np.float(np.size(bin_z))
    
    y_like, x_like = np.histogram(bin_z, bins=200, normed=True) # take a histogram for the redshift distribution 
    x_like = x_like[:-1]
            
    bin_sp = spl(x_like,y_like,s=0.0, ext=1)     #Interpolate 
    n_z = bin_sp(z)   # this is the normalized n(z) in the range (0.0,4.0)
    
    Fisher_RM += FSB(mean_z, bias, n_z, f_sky, N_gal)
```

Let's plot the constraints from Redmagic only

```python
matrix_RM = Fisher_RM
fisher_RM = fm.fisher_matrix( fisher_matrix=matrix_RM, fiducial = fiducial, param_names=param_names )
fisher_RM.name = 'RedMaGic'


fisher_list = fpa.CosmicFish_FisherAnalysis()
fisher_list.add_fisher_matrix([fisher_RM])
fisher_plotter = fp.CosmicFishPlotter(fishers=fisher_list)

fisher_plotter.new_plot()
fisher_plotter.plot_tri()
fisher_plotter.export('RedMaGic.pdf', dpi=100)
```
